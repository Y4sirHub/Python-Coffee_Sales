```python
# pip install matplotlib

import matplotlib.pyplot as plt
```

```python
df_coffee = pd.read_csv('/kaggle/input/coffee-sales/index.csv')
df_coffee
```

```python
df_coffee.info()
```

```python
unique_coffee_names = df_coffee['coffee_name'].unique()
print(unique_coffee_names)
```

```python
# Sample data
 
categories = df_coffee['coffee_name']
values = df_coffee['money']
 
# Create a bar plot
plt.figure(figsize=(10,6))
plt.bar(categories, values, color ='peru', width = 0.5, edgecolor = 'black')
plt.title('Amount generated per coffee variant',fontweight = 'bold', color = 'saddlebrown', fontsize = '15')

## Customisation

# Adding axis labels
plt.xlabel('Coffee Name', fontweight = 'bold', color = 'sienna', fontsize = '10')
plt.xticks(rotation = 45)
plt.ylabel('Amount', fontweight = 'bold', color = 'sienna', fontsize = '10')
plt.legend()

# Showing the plot
plt.show()
```

```python
# pip install seaborn

import seaborn as sns
import matplotlib.pyplot as plt
```

# Scatter Plot

```python
# Scatter Plot for Cash Type

plt.figure(figsize=(4,6))
sns.scatterplot(data=df_coffee, x='cash_type', y='money', hue='coffee_name',palette='coolwarm')
plt.legend(title='Payment method',title_fontsize='08', fontsize='06',bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
# Scatter Plot for Coffee Name

plt.figure(figsize=(12,8))
sns.scatterplot(data=df_coffee, x='coffee_name', y='money', hue='cash_type',palette='coolwarm')
plt.legend(title='Payment method',title_fontsize='08', fontsize='06',loc='lower left')
plt.show()
```

```python
sns.set_theme(style='white')
plt.figure(figsize=(4,6))
sns.scatterplot(data=df_coffee, x='cash_type', y='coffee_name', hue='money',palette='coolwarm')
#plt.legend(title='Payment method',title_fontsize='08', fontsize='06',bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
sns.set_theme(style="white")
plt.figure(figsize=(5,7))
sns.relplot(data=df_coffee, x='cash_type', y='coffee_name', size='money', hue='money', palette = 'coolwarm')
#plt.legend(title='Payment method',title_fontsize='08', fontsize='06',bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
sns.set_theme(style="dark")
plt.figure(figsize=(5,7))
sns.relplot(data=df_coffee, x='cash_type', y='coffee_name', size='money', hue='money', palette = 'coolwarm')
#plt.legend(title='Payment method',title_fontsize='08', fontsize='06',bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
